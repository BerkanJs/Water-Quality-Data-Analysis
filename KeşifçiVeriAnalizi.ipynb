#### Coefficient of Variation (CV)

```python
def calculate_cv(column):
    return (df[column].std() / df[column].mean()) * 100

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
numerical_columns=df.drop(columns=['Potability'])
cv_values = {col: calculate_cv(col) for col in numerical_columns}

for col, cv in cv_values.items():
    print(f"Coefficient of Variation for {col}: %{cv:.2f}")
    if cv > 25:
        print(f"  -> {col} verisi geniş bir dağılıma sahip, yayılma fazla!")
    else:
        print(f"  -> {col} verisi dar bir dağılıma sahip.")
```

#### MCAR (Little's Test)

- McAR değilse eksik veri, verinin eksik olmasının rastgele olmadığını, yani eksikliğin bir desen ya da ilişki taşıdığını gösterir

```python
import pandas as pd
from scipy import stats


df_MCAR = pd.DataFrame(df)


print("Eksik Veri Sayıları:\n", df_MCAR.isnull().sum())


p_values = {}

for column in df_MCAR.columns:

    df_no_missing = df_MCAR[df_MCAR[column].notna()]
    df_missing = df_MCAR[df_MCAR[column].isna()]

    if len(df_missing) == 0 or len(df_no_missing) == 0:
        continue  


    stat, p_normal = stats.shapiro(df_no_missing[column])
    if p_normal > 0.05:
        
        _, p_value = stats.ttest_ind(df_no_missing[column], df_missing[column], equal_var=False)
    else:
       
        _, p_value = stats.mannwhitneyu(df_no_missing[column], df_missing[column])

    p_values[column] = p_value
    print(f"{column} - p-değeri: {p_value}")


for column, p_value in p_values.items():
    if p_value > 0.05:
        print(f"{column} - Eksik veriler MCAR olarak kabul edilebilir (p > 0.05)")
    else:
        print(f"{column} - Eksik veriler MCAR değil (p < 0.05)")
```

- MCAR analizinin sonucu bize eksik verilerin tamamen rassal şekilde oluşmadığını söylüyor

##### MCAR analizinin sonucu bize eksik verinin tamamen rassal olusmadıgını bir desen oldugnu yani farklı sutunlardaki eksik verilerin birbirleriyle ilişkili olabileceğini söylüyor. Bu durumda eksik verileri doldururken MICE yöntemi median ile doldurmaya göre daha uygundur. Çünkü medyan ile doldurma diğer sütunların hedef sütunla olan ilişkisini dikkate almaz

#### MICE 

```python
from sklearn.experimental import enable_iterative_imputer  
from sklearn.impute import IterativeImputer


mice_imputer = IterativeImputer(max_iter=10, random_state=0)


df_imputed = pd.DataFrame(mice_imputer.fit_transform(df))


df_imputed.columns = df.columns


df_imputed.head(15)
```

```python
import pandas as pd


original_summary = df.describe().T[['mean', 'std']]


imputed_summary = df_imputed.describe().T[['mean', 'std']]


comparison = original_summary.join(imputed_summary, lsuffix='_original', rsuffix='_imputed')


print(comparison)
```
